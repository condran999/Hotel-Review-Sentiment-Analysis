# Web Scrapping Review from Trip advisor

```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
from time import sleep
from random import randint
import matplotlib.pylab as plt
```

## Web Scraping

```python
# Since the page number changes by 5 in the next url page
page = np.arange(5, 5001, 5)  
Name = []
Title = []
Review = []
Rating = []
for pages in page:
    # for loop to collect the Mutiple Pages as defined
    url = 'https://www.tripadvisor.in/Hotel_Review-g304554-d307116-Reviews-or' + str(
        pages
    ) + '-Grand_Hyatt_Mumbai_Hotel_Residences-Mumbai_Maharashtra.html#REVIEWS'

    # import the Url details to Python
    req = requests.get(url, headers={'User-Agent': 'Mozilla/5.0'})
    html = req.content

    # Put it in soup
    sleep(randint(2, 8))  
    soup = BeautifulSoup(html, 'html.parser')

    # for loop to extract Customer Names
    for a in soup.find_all('a', class_='ui_header_link _1r_My98y'):
        Name.append(a.get_text(strip=True))
        
    # for loop to extract Review Title
    for a in soup.find_all('div', class_='glasR4aX'):
        Title.append(a.get_text(strip=True))

    # for loop to extract Reviews
    for a in soup.find_all('q', class_='IRsGHoPm'):
        Review.append(a.get_text())

    # for loop to extract Review Rating
    for a in soup.find_all('div', class_='nf9vGX55'):
        # a.span['class'] is used to only select the span class text which has the rating ['ui_bubble_rating', 'bubble_50']
        Rating.append(a.span['class'])
```

```python
# Extraction of the Rating out of --> this format text ['ui_bubble_rating', 'bubble_50'] generated through above for loop
Final_Rating = []
for i in Rating:
    # select only 'bubble_50' for the list and converted to string to extarct the rating
    bubble_rating = str(i[1])
    # selected the 2nd last letter which is the rating and converted to integer
    Final_Rating.append(int(bubble_rating[-2]))
```

```python
# Data Frame created for teh details extracted 
df=pd.DataFrame({'Customer_name':Name,'Rating':Final_Rating,'Review_Title':Title,'Review':Review})
df
```

```python
# Saved the data frame in csv file
df.to_csv('Tripadvisor_Grand_Hyatt_Review.csv')
```
